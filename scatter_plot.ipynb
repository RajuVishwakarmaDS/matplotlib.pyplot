```python
import matplotlib.pyplot as plt

x=[1,3,5,2,4,6]
y=[2,4,6,1,3,5]
t=[1,2,3,4,5,6]

plt.scatter(x,t,color='r',label='label one',marker='X')
plt.scatter(y,t,color='b',label='label two',marker='o')

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title(' title ')

plt.grid()
plt.legend()
plt.show()
```
